```
import sys, os
os.chdir('/content/FCF')
sys.path.insert(0, '.')
import torch
print('CUDA:', torch.cuda.is_available())
print('CPU limit:', os.cpu_count(), 'cores')
torch.set_num_threads(min(os.cpu_count() or 2, 4))
```

```
import sys, os
os.chdir('/content/FCF')
sys.path.insert(0, '.')
from fcf.config import load_config
from fcf.primordial_layer import PrimordialLayer
from fcf.language_trainer import LanguageTrainer
from fcf.tokenizer_utils import load_tokenizer

tokenizer = load_tokenizer('tokenizer.json')
layer = PrimordialLayer(load_config())
trainer = LanguageTrainer(layer=layer, tokenizer=tokenizer, checkpoint_dir='checkpoints/language_wiki')
print(sum(p.numel() for p in layer.parameters()), 'params')
print('Tokenizer vocab:', tokenizer.get_vocab_size())
```

```
trainer.train(max_steps=20000, device='cpu', use_wikipedia=True)
```

```
import sys, os
os.chdir('/content/FCF')
sys.path.insert(0, '.')
from fcf.utils import save_primordial_layer

save_primordial_layer(layer, 'checkpoints/language_wiki/final')

!zip -q fcf_weights.zip checkpoints/language_wiki/final/weights.pt
from google.colab import files
files.download('fcf_weights.zip')
print('Weights downloaded. Done.')
```
